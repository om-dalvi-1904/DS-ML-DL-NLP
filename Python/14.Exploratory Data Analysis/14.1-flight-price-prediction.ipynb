```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
# get the basic info about the data

df.info()
```

```python
df.describe()  #returns only numerical feature
```

```python
# split the date into dd, mm, yyyy

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df
```

```python
df.info()
```

```python
# convert them into int
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.describe()
```

```python
# drop the date of journey column

df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(" ")[0])
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
# convert them into int
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
# drop the arrival time column

df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.head(1)
```

```python
df['Dept_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dept_Minute'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df.info()
```

```python
# convert them into int dept_time
df['Dept_Hour'] = df['Dept_Hour'].astype(int)
df['Dept_Minute'] = df['Dept_Minute'].astype(int)
```

```python
df.describe()
```

```python
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, np.nan: 1})
```

```python
df.head()
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df["Duration_Hour"] = df['Duration'].str.split(" ").str[0].str.split('h').str[0]
df["Duration_Minutes"] = df['Duration'].str.split(" ").str[0].str.split('m').str[0]

df.head(2)
```
