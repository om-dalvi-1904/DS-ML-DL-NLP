```python
import pandas as pd
```

```python
# series in pandas
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# create a series from dictionary
data = {'a':1, 'b':2,'c':3}
pd.Series(data)
```

```python
data = [10, 20, 30]
index = ['a', 'b', 'c']

pd.Series(data, index)
```

```python
# data frames in python
# create a data frame from a dictionary

data = {
    'Name': ['Krish', 'John', 'Jack'],
    'Age': [25, 32, 30],
    'City': ['Pune', 'Agra', 'Chennai'],
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```
